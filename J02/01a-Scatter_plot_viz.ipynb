```python
for column in df.columns:
    if df[column].dtype == 'object':
        print(f"{column} : Qualitative")
    elif df[column].dtype == 'int64' or df[column].dtype == 'float64':
        if df[column].nunique() < 10:
            print(f"{column} : Discrete Quantitative")
        else:
            print(f"{column} : Continuous Quantitative")
    else:
        print(f"{column} : Autre")
```

3. When creating a scatterplot, look at how the _LotArea_ column behaves in relation to the _SalePrices_ column.

```python
sns.scatterplot(x='SalePrice',y='LotArea',data = df)
```

4. Refine your visualization by keeping only those houses that have a _LotArea_ of less than 20,000 square feet and a price of less than $500,000.

```python
sns.scatterplot(x="LotArea", y="SalePrice", data= df.loc[(df['LotArea']<20000) & (df['SalePrice']<500000), :])
#sans creer de df.intermediaire mais sinon option correcte                
                
```

5. When creating a point cloud, look at the relationship between the _LotFrontage_ and the _LotArea_.
